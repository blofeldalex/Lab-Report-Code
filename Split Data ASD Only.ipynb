```python
import pandas as pd ## import the library
X = pd.read_csv(r"data_ASD_only.csv") ## read the csv into a dataframe 'X' (2d table - rows and columns)
X.head(10) ## display the first 10 rows
```

```python
X
```

```python
## Check start values of sex so we know what we're working with
print('Number of rows: ',len(X.index))
X['Sex'].value_counts()

## This shows 244 sex==0.0, 243 sex==1.0 and since 488 rows total, 1 row where sex==null
```

```python
## sex0 is a dataframe containing only the values from X where sex == 0.0
## sex1 only has Sex == 1.0 and the one null row
sex0 = X[X['Sex']==0.0]
sex1 = X[X['Sex']!=0.0]

## randomly shuffle both sex0 and sex1 to ensure random split
shuf0 = sex0.sample(frac=1)
shuf1 = sex1.sample(frac=1)

## halfLengthShuf0 is equal to the number of rows in shuf0 divided by 2 (rounded to int)


halfLengthShuf0 = int(0.5 * len(shuf0))
halfLengthShuf1 = int(0.5 * len(shuf0))

## We then use the half length int to split shuf0 in half using indexing. 
## first_half_shuf0 = the first half of shuf0
first_half_shuf0 = shuf0[ : halfLengthShuf0]
first_half_shuf1 = shuf1[ : halfLengthShuf1]
second_half_shuf0 = shuf0[halfLengthShuf0 : ]
second_half_shuf1 = shuf1[halfLengthShuf1 : ]


## We then concatenate one half of the randomly shuffled sex==0.0 rows 
## with one half of the randomly shuffled sex==1.0 rows
## and concatenate the remaining two halves together also
split0 = pd.concat([first_half_shuf0 , first_half_shuf1])
split1 = pd.concat([second_half_shuf0 , second_half_shuf1])


## Do it all at once instead
#split0 = pd.concat([shuf0[:int(0.5 * len(shuf0))], shuf1[:int(0.5 * len(shuf0))]])
#split1 = pd.concat([shuf0[int(0.5 * len(shuf0)):], shuf1[int(0.5 * len(shuf0)):]])


## Reshuffle split0 and split1 to ensure that everyone is sititng boy-girl-boy-girl (approximately)
split0 = split0.sample(frac=1)
split1 = split1.sample(frac=1)
```
